## Luku 12.2.3 Tehtävä 1

X ~ N(μ, 25)    -> σ = 5 tiedossa

otos (18 kpl):
97.0  103.3  96.2  93.8  103.9  96.4
101.5 104.1  107.7 100.3  93.4  98.5
102.1  98.6  104.8 103.7  97.3  99.2

Tehtävä: laske 95%, 99% ja 99.9% luottamusvälit odotusarvolle μ

```python
import numpy as np
from scipy import stats

data = np.array([
    97.0, 103.3, 96.2, 93.8, 103.9, 96.4,
    101.5, 104.1, 107.7, 100.3, 93.4, 98.5,
    102.1, 98.6, 104.8, 103.7, 97.3, 99.2
])

n = len(data)
keskiarvo = np.mean(data)
se = 5 / np.sqrt(n)

print("otoskoko n =", n)
print("keskiarvo   =", round(keskiarvo, 3))
print("se (virhe)  =", round(se, 4))
print()

# z-arvot (95%, 99%, 99.9%)
z95  = stats.norm.ppf(0.975)     # ≈ 1.960
z99  = stats.norm.ppf(0.995)     # ≈ 2.576
z999 = stats.norm.ppf(0.9995)    # ≈ 3.291

print("95%  luottamusväli:", round(keskiarvo - z95 * se, 2), "–", round(keskiarvo + z95 * se, 2))
print("99%  luottamusväli:", round(keskiarvo - z99 * se, 2), "–", round(keskiarvo + z99 * se, 2))
print("99.9% luottamusväli:", round(keskiarvo - z999 * se, 2), "–", round(keskiarvo + z999 * se, 2))
```

## Luku 12.2.3 Tehtävä 4

X ~ N(μ, σ²)    molemmat tuntemattomia

otos (20 kpl):
44.6  47.2  61.9  52.9  36.9
55.2  50.6  57.7  44.5  63.5
48.9  61.2  50.5  45.3  59.9
43.9  48.0  51.5  50.3  54.0

Tehtävä: 95%, 99% ja 99.9% luottamusvälit odotusarvolle μ
(käytetään t-jakaumaa koska σ tuntematon)

```python
import numpy as np
from scipy import stats

data = np.array([
    44.6, 47.2, 61.9, 52.9, 36.9,
    55.2, 50.6, 57.7, 44.5, 63.5,
    48.9, 61.2, 50.5, 45.3, 59.9,
    43.9, 48.0, 51.5, 50.3, 54.0
])

n = len(data)
keskiarvo = np.mean(data)
otos_sd = np.std(data, ddof=1)          # otoshajonta (n-1)
se = otos_sd / np.sqrt(n)

print("otoskoko n       =", n)
print("keskiarvo        =", round(keskiarvo, 3))
print("otos_sd (s)      =", round(otos_sd, 3))
print("se (virhe)       =", round(se, 4))
print()

# vapaudenasteet
df = n - 1

# t-arvot (kaksisuuntainen)
t95  = stats.t.ppf(0.975,  df)    # ≈ 2.093
t99  = stats.t.ppf(0.995,  df)    # ≈ 2.861
t999 = stats.t.ppf(0.9995, df)    # ≈ 3.883

print("95%  luottamusväli:", round(keskiarvo - t95  * se, 2), "–", round(keskiarvo + t95  * se, 2))
print("99%  luottamusväli:", round(keskiarvo - t99  * se, 2), "–", round(keskiarvo + t99  * se, 2))
print("99.9% luottamusväli:", round(keskiarvo - t999 * se, 2), "–", round(keskiarvo + t999 * se, 2))
```

## Luku 12.2.3 Tehtävä 6

X ~ N(μ, σ²)    molemmat tuntemattomia

otos (27 kpl):
17.6  21.4  21.5  19.3  20.7  18.1  23.1
20.2  21.5  23.8  21.0  16.9  19.3  20.0
19.1  20.0  16.5  18.8  19.1  17.0  17.6
20.4  17.0  18.5  24.4  20.4  18.3

Tehtävä:
- laske 99 % luottamusvälin pituus tällä otoksella
- arvioi montako havaintoa tarvitaan jotta 99 % välin pituus on enintään 0.4

```python
import numpy as np
from scipy import stats

data = np.array([
    17.6, 21.4, 21.5, 19.3, 20.7, 18.1, 23.1,
    20.2, 21.5, 23.8, 21.0, 16.9, 19.3, 20.0,
    19.1, 20.0, 16.5, 18.8, 19.1, 17.0, 17.6,
    20.4, 17.0, 18.5, 24.4, 20.4, 18.3
])

n = len(data)
keskiarvo = np.mean(data)
s = np.std(data, ddof=1)          # otoshajonta
se = s / np.sqrt(n)

print("n =", n)
print("keskiarvo =", round(keskiarvo, 3))
print("s =", round(s, 3))
print("se =", round(se, 4))

df = n - 1
t99 = stats.t.ppf(0.995, df)

puoli_leveys = t99 * se
pituus = 2 * puoli_leveys

print("t99 (df=" + str(df) + ") =", round(t99, 4))
print("99 % välin pituus =", round(pituus, 4))
```

```python
# arvio tarvittavasta otoskoko n:stä jotta pituus max 0.4

# käytetään samaa s-arvoa approksimaatioon
# kun n iso -> t ≈ z = 2.576 (99 %)

z99 = 2.576
max_pituus = 0.4
puoli_max = max_pituus / 2     # 0.2

# tarvittava se = puoli_max / z
se_tarve = puoli_max / z99

# n = (s / se_tarve)^2
n_tarve = (s / se_tarve) ** 2

print("arvio n tarvittava (z-approx):", round(n_tarve))
print("pyöristetty ylöspäin:", np.ceil(n_tarve).astype(int))
```

## Luku 12.2.3 Tehtävä 9**

Tehtävä 4 otos (n=20) varianssi ja sd 99 % väli
Tehtävä 6 otos (n=27) varianssi ja sd 99 % väli

käytetään chi²-jakaumaa koska σ tuntematon

```python
import scipy.stats
print("scipy ok")
```

```python
import numpy as np
from scipy import stats

# Tehtävä 4 data (n=20)
data4 = np.array([44.6, 47.2, 61.9, 52.9, 36.9, 55.2, 50.6, 57.7, 44.5, 63.5,
                  48.9, 61.2, 50.5, 45.3, 59.9, 43.9, 48.0, 51.5, 50.3, 54.0])

n4 = len(data4)
s2_4 = np.var(data4, ddof=1)          # otosvarianssi
s4 = np.sqrt(s2_4)

print("Tehtävä 4")
print("n =", n4)
print("otosvar s² =", round(s2_4, 3))
print("otos_sd s =", round(s4, 3))

df = n4 - 1
chi_low  = stats.chi2.ppf(0.005,  df)   # 99 % -> 0.5 % + 0.5 %
chi_high = stats.chi2.ppf(0.995, df)

var_low  = (df * s2_4) / chi_high
var_high = (df * s2_4) / chi_low

sd_low  = np.sqrt(var_low)
sd_high = np.sqrt(var_high)

print("99 % varianssille:", round(var_low, 3), "–", round(var_high, 3))
print("99 % keskihajonnalle:", round(sd_low, 3), "–", round(sd_high, 3))
print()
```

```python
# Tehtävä 6 data (n=27)
data6 = np.array([17.6, 21.4, 21.5, 19.3, 20.7, 18.1, 23.1, 20.2, 21.5,
                  23.8, 21.0, 16.9, 19.3, 20.0, 19.1, 20.0, 16.5, 18.8,
                  19.1, 17.0, 17.6, 20.4, 17.0, 18.5, 24.4, 20.4, 18.3])

n6 = len(data6)
s2_6 = np.var(data6, ddof=1)
s6 = np.sqrt(s2_6)

print("Tehtävä 6")
print("n =", n6)
print("otosvar s² =", round(s2_6, 3))
print("otos_sd s =", round(s6, 3))

df = n6 - 1
chi_low  = stats.chi2.ppf(0.005,  df)
chi_high = stats.chi2.ppf(0.995, df)

var_low  = (df * s2_6) / chi_high
var_high = (df * s2_6) / chi_low

sd_low  = np.sqrt(var_low)
sd_high = np.sqrt(var_high)

print("99 % varianssille:", round(var_low, 3), "–", round(var_high, 3))
print("99 % keskihajonnalle:", round(sd_low, 3), "–", round(sd_high, 3))
```
